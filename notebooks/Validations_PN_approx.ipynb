Also, we can talk about the values tabulated in Table 3 via calculating r at f = 20 Hz and taking the slice in the plot at that binary separation

```python
solar_mass = 4.925490947671582e-06
def r_at_20_Hz(mcz_in_solar, q, f_ref = 20.0):
    eta = q/(1+q)**2
    total_mass = (eta**(-0.6))*mcz_in_solar*solar_mass
    r20 = ((np.pi*f_ref*total_mass)**(-2/3))
    return r20
```
